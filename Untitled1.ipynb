```python
import pandas as pd
# Load your dataset
df = pd.read_csv('disney movie total gross.csv')

# See the first 5 rows
df.head()
```

```python
df.info()
```

```python
# 1. Convert 'Date Released' to proper date format
df['Date Released'] = pd.to_datetime(df['Date Released'], dayfirst=True)

# 2. Create a new 'Year' column from the release date
df['Year'] = df['Date Released'].dt.year

# 3. Check your cleaned data
df.head()
```

```python
import matplotlib.pyplot as plt

# Count number of movies released each year
movies_per_year = df['Year'].value_counts().sort_index()

# Plot
plt.figure(figsize=(14, 6))
movies_per_year.plot(kind='bar')
plt.title('Number of Disney Movies Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Count how many movies fall into each genre
genre_counts = df['Genre'].value_counts()

# Plot it as a bar chart
plt.figure(figsize=(10, 6))
genre_counts.plot(kind='bar')
plt.title('Number of Disney Movies by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Group by genre and calculate average inflation-adjusted gross
revenue_by_genre = df.groupby('Genre')['Inflation Adjusted Gross'].mean().sort_values(ascending=False)

# Plot the result
plt.figure(figsize=(12, 6))
revenue_by_genre.plot(kind='bar')
plt.title('Average Inflation-Adjusted Revenue by Genre')
plt.xlabel('Genre')
plt.ylabel('Avg. Revenue ($)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Group by year and calculate average inflation-adjusted revenue
avg_revenue_by_year = df.groupby('Year')['Inflation Adjusted Gross'].mean()

# Plot it
plt.figure(figsize=(14, 6))
avg_revenue_by_year.plot(kind='line', marker='o')
plt.title('Average Disney Movie Revenue by Year (Inflation Adjusted)')
plt.xlabel('Year')
plt.ylabel('Avg. Revenue ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Sort the DataFrame by inflation-adjusted revenue in descending order
top_10_movies = df.sort_values(by='Inflation Adjusted Gross', ascending=False).head(10)

# Display the top 10
top_10_movies[['Movie Title', 'Year', 'Inflation Adjusted Gross']]
```

```python
# Plot top 10 movies
plt.figure(figsize=(12, 6))
plt.barh(top_10_movies['Movie Title'], top_10_movies['Inflation Adjusted Gross'], color='skyblue')
plt.xlabel('Inflation Adjusted Gross ($)')
plt.title('Top 10 Highest-Grossing Disney Movies (Adjusted for Inflation)')
plt.gca().invert_yaxis()  # So the highest movie is at the top
plt.tight_layout()
plt.show()

```

```python
# Create a new 'Decade' column
df['Decade'] = (df['Year'] // 10) * 10  # e.g., 1994 → 1990

# Group by decade and get average inflation-adjusted gross
revenue_by_decade = df.groupby('Decade')['Inflation Adjusted Gross'].mean()

# Plot it
plt.figure(figsize=(10, 6))
revenue_by_decade.plot(kind='bar', color='orange')
plt.title('Average Disney Movie Revenue by Decade (Inflation Adjusted)')
plt.xlabel('Decade')
plt.ylabel('Avg. Revenue ($)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Group by MPAA rating and calculate average adjusted gross
rating_revenue = df.groupby('MPAA Rating')['Inflation Adjusted Gross'].mean().sort_values(ascending=False)

# Plot
plt.figure(figsize=(8, 5))
rating_revenue.plot(kind='bar', color='green')
plt.title('Average Revenue by MPAA Rating (Inflation Adjusted)')
plt.xlabel('MPAA Rating')
plt.ylabel('Avg. Revenue ($)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
